```python
class Parent():
  def myMethod(self):
    print('부모메쏘드')

class Child(Parent):  # 재정의(overriding): 자식함수가 우선
  def myMethod(self):
    print('자식메쏘드')

ch = Child()  # 인스턴스
ch.myMethod()
```

```python
class Parent:                  # 부모클래스
  parentAttr = 100             # 공유변수인 크랠스 변수
  def __init__(self, name):    # 부모 생성자
    self.name = name           # 멤버 변수
    print("부모 생성자")
  def parentMethod(self):
    print("부모 메써드",self.name)
  def setAttr(self, attr):
    Parent.parentAttr = attr
  def getAttr(self):
    print("부모 속성", Parent.parentAttr)
class Child(Parent):     # 상속받은 자식클래스
  def __init__(self):    # 자식 생성자
    Parent.__init__(self,"서울") # 부모의 생성자 명시적 호출
    print("자식 생성자")
  def childMethod(self):
    print("자식 메써드")
  def print_attribute(self):
    self.setAttr(500)
    print("자식 호출된 부모 속성", Parent.parentAttr)
```

```python
c = Child()
c.childMethod()
c.parentMethod()
c.setAttr(200)
print(c.getAttr())
c.print_attribute()
c.parentAttr
```

# 다음 Point 클래스를 상속받은 3D 클래스를 만들어보시요

```python
import math
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y
  def translate(self, dx, dy):
    self.x += dx
    self.y += dy
```

```python
class Point3D(Point):
  def __init__(self, x, y, z):
    Point.__init__(self, x, y)
    self.z=z
  def translate(self, dx, dy, dz):
      Point.translate(self, dx, dy)
      self.z += dz

```

```python
class Shape:
  def __init__(self):
    print("부모 클래스")
  def area(self):
    print("면적")
  def perimeter(self):
    print("둘레")
```

```python
class Triangle(Shape):
  def __init__(self, width, heidth):
    self.width = width
    self.heidth = heidth
  def area(self):
    print("삼각형의 넓이는= ", self.width * self.heidth/2)
  def perimeter(self):
    print("삼각형의 둘레는= ", self.width + math.sqrt(math.pow(self.width/2,2) + math.pow(self.heidth,2))*2)
```

```python
# Circle(point, radius), Rectangle(point) 클래스 2개로 구현
class Circle(Shape):
  def __init__(self, center, radius):
    self.center = center
    self.radius = radius
  def area(self):
    print("원의 넓이는 = ", math.pi * self.radius**2)
  def perimeter(self):
    print("원의 둘레는 = ", 2*math.pi*self.radius)

class Ractangle(Shape):
  def __init__(self, po1, po2):
    self.po1 = po1
    self.po2 = po2
  def area(self):
    print("사각형의 넓이는= ", abs(self.po1.x-self.po2.x)* abs(self.po1.y-self.po2.y))
  def perimeter(self):
    print("사각형의 둘레는=", abs(self.po1.x-self.po2.x)*2 + abs(self.po1.y - self.po2.y)*2)

```

```python
import random
shape_list=[]
for i in range(500):
  rand = random.randint(0,10)
  if rand%3 == 0:
    width = random.randint(1,10)
    heidth = random.randint(1,10)
    tri = Triangle(width,heidth)
    shape_list.append(tri)
  elif rand%3 == 1:
    x = random.randint(1,10)
    y = random.randint(1,10)
    rect = Ractangle(Point(10,20), Point(x,y))
    shape_list.append(rect)
  else:
    r = random.randint(1,10)
    ci = Circle(Point(10,20),r)
    shape_list.append(ci)

for i in range(500):
  shape_list[i].area()
  shape_list[i].perimeter()
```

```python
class Evaluation:
  def printcalcGrade(self, average):
    pass
class Hakjum(Evaluation):
  def printcalcGrade(self, average):
    if average >= 90: return "A"
    elif average >= 80: return "B"
    elif average >= 70: return "C"
    elif average >= 60: return "D"
    else: return "F"
class Pass_Fail(Evaluation):
  def printcalcGrade(self, average):
    if average >= 60: return "합격"
    else:             return "불합격"
class Element_Evaluation(Evaluation):
  def printcalcGrade(self, average):
    if average >= 90: return "수"
    elif average >= 80: return "우"
    elif average >= 70: return "미"
    elif average >= 60: return "양"
    else: return "가"

# 인터페이스
class Evaluation_print:
  def print_evaluation(self,evaluation):
    jumsu = eval(input("점수를 입력하시요"))
    print(evaluation.printcalcGrade(jumsu))
```

```python
# (2,4) (3,7)
# vector의 요소 연산
class Vector :
  def __init__(self, x, y):
    self.x = x
    self.y = y
  def __repr__(self):
    return f"(좌표는 {self.x}, {self.y})"
  def __add__(self, other):
    return Vector(self.x + other.x, self.y + other.y)
  def __mul__(self, other):
    return Vector(self.x * other.x, self.y * other.y)
  def __sub__(self, other):
    return Vector(self.x - other.x, self.y - other.y)
  def __truediv__(self, other):
    return Vector(self.x/other.x , self.y/other.y)
  def __floordiv__(self, other):
    return Vector(self.x/other.x, self.y/other.y)
```

```python
# 문제 : 위 vector클래스를 3차원으로 확장하시요

class Vector :
  def __init__(self, x, y, z):
    self.x = x
    self.y = y
    self.z = z
  def __repr__(self):
    return f"(좌표는 {self.x}, {self.y}, {self.z})"
  def __add__(self, other):
    return Vector(self.x + other.x, self.y + other.y, self.z + other.z)
  def __mul__(self, other):
    return Vector(self.x * other.x, self.y * other.y, self.z * other.z)
  def __sub__(self, other):
    return Vector(self.x - other.x, self.y - other.y, self.z - other.z)
  def __truediv__(self, other):
    return Vector(self.x/other.x , self.y/other.y, self.z / other.z)
  def __floordiv__(self, other):
    return Vector(self.x/other.x, self.y/other.y, self.z / other.z)

  def norm(self): # Vector의 크기값(피타고라스의 정의)
    return math.sqrt(self.x**2 + self.y**2 + self.z**2)
  def normalize(self): # 정규화 => 사이즈1 => 방향
    norm = self.norm()
    return Vector(self.x/norm, self.y/norm, self.z/norm)
  def dot(self, other): # 내적: 크기값이 고려된 사이각
    return self.x * other.x + self.y * other.y + self.z * other.z
  def costheta(self, other): # cos(theta)
    dot = self.dot(other)
    norm1 = self.norm()
    norm2 = other.norm()
    return dot / ( norm1 * norm2 )
  def theta(self, other): # 라디안 -> (180 /math.pi)
    costheta = self.costheta(other)
    # 각도에서 cos(theta)의 값은 -1~1 사이의 값
    costheta = max(-1, min(1, costheta))
    print(costheta)
    return math.acos(costheta) * float( 180/math.pi)
```

```python
class Vector:
  def __init__(self,vec):
    self.vec = vec
  def __repr__(self):
    return f"데이터는 {self.vec}"
  def __add__(self, other):
    return Vector([self.vec[i] + other.vec[i] for i in range(len(self.vec))])
  def __mul__(self, other):
    return Vector([self.vec[i] * other.vec[i] for i in range(len(self.vec))])
  def __sub__(self, other):
    return Vector([self.vec[i] - other.vec[i] for i in range(len(self.vec))])
  def __truediv__(self, other):
    return Vector([self.vec[i] / other.vec[i] for i in range(len(self.vec))])
  def norm(self):
    return math.sqrt(sum([element * element for element in self.vec]))
  def normalize(self):
    norm = self.norm()
    return Vector([element/norm for element in self.vec])
  def dot(self, other):
    return sum([self.vec[i] * other.vec[i] for i in range(len(self.vec))])
  def costheta(self, other):
    dot = self.dot(other)
    norm1 = self.norm()
    norm2 = other.norm()
    return dot/(norm1 * norm2)
  def theta(self, other):
    costheta = self.costheta(other)
    costheta = max(-1, min(1, costheta))
    return math.acos(costheta) * (180/math.pi) # 라디안 가도를 각도로 변환

# 내적값은 크기를 고려한 사이각
# 내적 값의 사이각이 좁으면 같은 방향
```

```python
class Student:
  def __init__(self, evaluate, name='', kor=0, eng=0, mat=0, average=0, total=0, grade=0 ):
    self.name = name
    self.kor = kor
    self.eng = eng
    self.mat = mat
    self.total = total
    self.average = average
    self.grade = grade
    self.evaluate = evaluate
  def calc_total(self):
    return self.kor + self.mat + self.eng
  def calc_average(self):
    return round(self.total / 3.0, 2)
  def calc_grade(self, jumsu_point):
    if jumsu_point>=90:
      return "A"
    elif jumsu_point>80:
      return "B"
    elif jumsu_point>70:
      return "C"
    elif jumsu_point>60:
      return "D"
    else: return "F"
  def calc_extern_grade(self):
    evaluation = self.evaluate.printcalcGrade(self.average)
    return evaluation
  def __repr__(self):
      return f"국어 : {self.kor}, 영어: {self.eng}, 수학:{self.mat}, 총점: {self.total}, 평균: {self.average}, 학점:{self.grade} "
  def input_data(self) :
    self.name = input("이름을 입력하시요")
    self.kor = int(input("국어점수"))
    self.mat = int(input("수학점수"))
    self.eng = int(input("영어점수"))
    self.total = self.calc_total()
    self.average = self.calc_average()
    self.grade = self.calc_grade(self.average)
    self.grade = self.calc_extern_grade()
```
